```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Conectar ao banco de dados SQLite (substitua 'seu_banco.db' pelo nome do seu arquivo SQLite)
conn = sqlite3.connect('db.sqlite3')

# Escrever a query para selecionar as colunas desejadas
query = "SELECT tipo_de_oferta, categorias FROM anuncio_anuncio"

# Carregar os dados em um DataFrame
df = pd.read_sql_query(query, conn)

# Fechar a conexão
conn.close()

# Exibir o DataFrame
display(df)
```

```python
# Verificar os dados
print(df.head())
```

```python
# Contar o número de ocorrências de cada tipo_de_oferta por categoria
contagem_por_categoria = df.groupby(['categorias', 'tipo_de_oferta']).size().unstack().fillna(0)
```

```python
# Exibir a contagem
display(contagem_por_categoria)
```

```python
# Criar o gráfico
contagem_por_categoria.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.xlabel('Categorias')
plt.ylabel('Número de Ofertas')
plt.title('Distribuição de Tipo de Oferta por Categoria')
plt.xticks(rotation=45, ha='right')  # Rotacionar os rótulos no eixo X para melhor visualização
plt.legend(title='Tipo de Oferta')
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

O objetivo desta análise é entender a distribuição dos tipos de oferta (Contratante e Contratado) dentro das diferentes categorias no banco de dados. Essa análise visa identificar como cada tipo de oferta está distribuído entre as categorias e fornecer uma visão geral das tendências e padrões.
